#T-statistic is to reject or not reject a Null hypothesis for one or two numerical sample data set. It's actually standardized sample mean variable (each sample mean centered based on average of population, and divided by estimated sigma of sample mean variable (which is estimated based on the sigma of sample "data" (not mean) and the size of sample)) 
* for one sample dataset of size n, null hypo is if the sample mean is independant of the population mean parameter, mu (is not a representative of population). 
t=(xbar-mu)/sqrt(s^2/n)
xbar: mean of the sample
mu: population mean parameter
s: sample standard deviation
n: number of points in the sample data set
* for two sample data sets, null hypothesis is if the two samples are independant (not equal). t=(xbar1-xbar2)/sqrt[(s1^2/n1)+(s2/n2)]

#Chi2 statistic is to create a crosstab for two categorical features (variables) in a sample data set and determines if the Null hypothesis (that the two variables are independant from eachother) can be rejected.
\begin{align}
\chi^2 = \sum \frac{(observed_{ij} -expected_{ij})^2}{(expected_{ij})}
\end{align}
* observed_ij: observed frequency for that combination state(element) in the crosstab (row i column j)
expected_ij : expected frequency for that element of crosstab. It is calculated as expected_ij=(total #of observations with feature i)*(total # of observations with feature j)/total of observations(samples in the data set)

# Increasing the sample size reduces the std of sample means distribution. And makes it closer to a normal distribution (central limit theorem)

```
sample_means = []
for x in range(0,3000):
  coinflips = np.random.binomial(n=1, p=.5, size=500)
  one_sample = coinflips
  sample_means.append(coinflips.mean())

print(len(sample_means))
print(sample_means)
```

```
plt.hist(sample_means, bins=80)
plt.show()
```

What does the Central Limit Theorem State? That no matter the initial distribution of the population (binomial in this example), the distribution of sample means taken (or any other estimate of the sample data sets, such as std) will approximate a normal distribution as $n \rightarrow \infty$.

This has very important implications for hypothesis testing and is precisely the reason why the t-distribution begins to approximate the normal distribution as our sample size increases. In a lame way we could assume the t-test distribution is the distribution of sample means with their number of its sigma set by t-statistic

```
sample_means_small = []
sample_means_large = []
for x in range(0,3000):
  coinflips_small = np.random.binomial(n=1, p=.5, size=50)
  coinflips_large = np.random.binomial(n=1, p=.5, size=200)
  one_small_sample = coinflips_small
  one_small_large = coinflips_large
  sample_means_small.append(coinflips_small.mean())
  sample_means_large.append(coinflips_large.mean())

print(len(sample_means_small))
print(sample_means_small)
```

### NumPy by DEFAULT calculates the POPULATION standard deviation and NOT the SAMPLE standard deviation

We need to pass the parameter `ddof=1` in order to ensure that we're really calculating the sample standard deviation and not the population standard deviation (population has N and sample has N-1 in std denominator formula)




```
from scipy import stats

def confidence_interval(sample_data, confidence=.95):
  # Make sure we're working with a numpy array, flexibility of the function argument
  sample = np.array(sample_data)
  # Sample Mean
  x_bar = sample.mean()
  # Sample Standard Deviation
  s = np.std(sample, ddof=1)
  # Sample Size
  n = len(sample)
  # T statistic - correspond to our confidence level, and our degrees of freedom
  # first argument is the confidence value of one side of t-test, distribution.
  # for the case of 95% is going to be 1-5%/2=0.975 on each side
  t = stats.t.ppf((1+confidence) / 2, n-1)
  # Standard Error
  std_err = s / np.sqrt(n)
  # Margin of Error
  moe = std_err * t
  # Lower and Upper Bound
  upper = x_bar + moe
  lower = x_bar - moe

  return (lower, x_bar, upper)

confidence_interval(sample1['age'])
```

```
# Taking pandas series instead of numpy array
# Pandas series std method default for sample (N-1).
# numpy std method defaults for population (N)
def confidence_interval(data, confidence=.95):
  # Make sure we're working with a numpy array
  sample = pd.Series(data)
  # Sample Mean
  x_bar = sample.mean()
  # Sample Standard Deviation
  s = sample.std()
  # Sample Size
  n = len(sample)
  # T statistic - correspond to our confidence level, and our degrees of freedom
  t = stats.t.ppf((1+confidence) / 2, n-1)
  # Standard Error
  std_err = s / np.sqrt(n)
  # Margin of Error
  moe = std_err * t
  # Lower and Upper Bound
  upper = x_bar + moe
  lower = x_bar - moe

  return (lower, x_bar, upper)

confidence_interval(sample1['age'])
```

** Basically it say for that size of sample (n), with 95% confidence 
the sample mean vatiables falls withing the upper bound and lower bound of confidence level.

Steps:
1) take a sample of the population
2) create a variable names sample mean
3) estimate the std of sample mean based on the std of sample data and size of sample (std_sample_mean=std_sample date/sqrt(n))
4) based on the degree of freedom (n-1) and p-val (probability value) get the centered normalized value of the sample mean variable (.mean(sample_mean)-population average)/.std(sample_mean)
5) determine the confidence level based on sample mean and t -statistic and estimated std of sample mean (.mean(sample_mean)+/- t_stat*std_sample_mean)

## Relationship between Confidence Intervals and T-tests

### (Bonus Content) You will not be assessed on this material

Confidence Interval == Bounds of statistical significance for our t-test

A sample mean that falls inside of our confidence interval will "FAIL TO REJECT" our null hypothesis

A sample mean that falls outside of our confidence interval will "REJECT" our null hypothesis

```
confidence_interval(sample1['age'])
```

### Upper Bound

```
# inside the confidence interval
stats.ttest_1samp(sample1['age'], 44.36515)
```

```
# outside the confidence interval
stats.ttest_1samp(sample1['age'], 44.36516)
```

### Lower Bound

```
# inside the confidence interval
stats.ttest_1samp(sample1['age'], 32.33485)
```

```
# outside the confidence interval
stats.ttest_1samp(sample1['age'], 32.33483)
```

If a null hypothesis value falls outside of our confidence interval, we will reject the null hypothesis.

If a null hypothesis value falls inside of our confidence interval we will fail to reject the null hypothesis.

A null hypothesis that's just inside of our confidence interval == fail to reject



A null hypothesis that's just outside of our confidence interval == reject
